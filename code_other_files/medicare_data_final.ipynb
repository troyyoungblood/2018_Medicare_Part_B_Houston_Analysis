```python
# Store csv into df

# shortened file name to keep from getting file not found errors (mistyping the really long file name)

csv_file= "medicare_data_CY2018.csv" 
raw_med_df= pd.read_csv(csv_file, low_memory=False)
raw_med_df.head()
```

```python
raw_med_rename_df = raw_med_df.rename(columns={'National Provider Identifier': 'National_Provider_ID', 'Last Name/Organization Name of the Provider': 'Provider_Last_Name_Org',
                     "First Name of the Provider" : "First_Name", "Middle Initial of the Provider" : "MI", 
                     "Credentials of the Provider" : "Credentials", "Gender of the Provider" : "Gender", 
                     "Entity Type of the Provider" : "Entity_Type", "Street Address 1 of the Provider" : "Address_1_Provider",
                     "Street Address 2 of the Provider" : "Address_2_Provider", "City of the Provider" : "City_Provider", 
                     "Zip Code of the Provider" : "Zip_Code", "State Code of the Provider" : "State_Code", 
                     "Country Code of the Provider" : "Country_Code", "Provider Type" : "Provider_Type", 
                     "Medicare Participation Indicator" : "Medicare_Part_Indicator", "Place of Service" : "Place_of_Service", 
                     "HCPCS Code" : "HCPCS_Code", "HCPCS Description" : "HCPCS_Description",
                     "HCPCS Drug Indicator" : "HCPCS_Drug_Indicator", "Number of Services" : "Num_Services", 
                     "Number of Medicare Beneficiaries" : "Number_Medicare_Ben", "Number of Distinct Medicare Beneficiary/Per Day Services" : "Number_Distinct_Ben_Per_Day", 
                     "Average Medicare Allowed Amount" : "Avg_Med_Allowed", "Average Submitted Charge Amount" : "Avg_Submitted_Charge",
                     "Average Medicare Payment Amount" : "Avg_Med_Payment", "Average Medicare Standardized Amount" : "Avg_Med_Stand"                     
                     })
```

```python
tx_raw_df = raw_med_rename_df.loc[raw_med_rename_df['State_Code'] == 'TX']
tx_raw_df.head()
```

```python
hou_raw_df = tx_raw_df.loc[tx_raw_df['City_Provider'] == 'HOUSTON']
hou_raw_df.head()
```

```python
hou_raw_df.to_csv ('houston_raw.csv', index = False)
```
